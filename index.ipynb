_______________________________________________________________________________________________________________________________

A fault-tolerant system is capable of providing uninterrupted service despite one of its components failing.  A RDD distributes the data acrosse multiple servers and includes backup of the data.







## Question 5

What is MapReduce? How is it related to Hadoop?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

MapReduce is the concept of having multiple servers run a query or function at the same time and then combine the results.  Running a query on multiple servers is the "map" step and combining the results from each server is the "reduce" step.  MapReduce is used to process huge datasets and Hadoop is a collection of the frameworks.
